## **Phần 1: NumPy cơ bản**

### **Bài tập 1: Tạo mảng và thao tác cơ bản**

1 Tạo một mảng NumPy với các giá trị từ 1 đến 20.\
2 Tìm tổng, giá trị lớn nhất, nhỏ nhất và trung bình của mảng.\
3 Tạo một mảng 2D (3x5) chứa các số ngẫu nhiên từ 0 đến 100.\
4 Lấy hàng thứ 2 và cột thứ 3 của mảng 2D.

```python

import numpy as np

# 1. Tạo một mảng NumPy với các giá trị từ 1 đến 20
arr = np.arange(1, 21)
print("Mảng từ 1 đến 20:\n", arr)

# 2. Tìm tổng, giá trị lớn nhất, nhỏ nhất và trung bình của mảng
print("Tổng của mảng:", arr.sum())
print("Giá trị lớn nhất:", arr.max())
print("Giá trị nhỏ nhất:", arr.min())
print("Giá trị trung bình:", arr.mean())

# 3. Tạo một mảng 2D (3x5) chứa các số ngẫu nhiên từ 0 đến 100
arr_2d = np.random.randint(0, 101, (3, 5))
print("Mảng 2D ngẫu nhiên:\n", arr_2d)

# 4. Lấy hàng thứ 2 và cột thứ 3 của mảng 2D
row_2 = arr_2d[1, :]  # Lấy hàng thứ 2 (chỉ số 1)
col_3 = arr_2d[:, 2]  # Lấy cột thứ 3 (chỉ số 2)

print("Hàng thứ 2:", row_2)
print("Cột thứ 3:", col_3)
```

### **Bài tập 2: Các thao tác nâng cao**

1 Tạo một mảng NumPy chứa 20 giá trị ngẫu nhiên từ 0 đến 1.\
2 Chuẩn hóa mảng này (đưa các giá trị về khoảng [0, 1]).\
3 Tính tích vô hướng (dot product) của hai mảng 1D: [1, 2, 3] và [4, 5,6].\
4 Tạo một ma trận 5x5 và tính định thức (determinant) và nghịch đảo của ma trận.

```python

import numpy as np

# 1. Tạo một mảng NumPy chứa 20 giá trị ngẫu nhiên từ 0 đến 1
arr = np.random.rand(20)
print("Mảng ngẫu nhiên từ 0 đến 1:\n", arr)

# 2. Chuẩn hóa mảng (đưa các giá trị về khoảng [0, 1])
arr_min = arr.min()
arr_max = arr.max()
arr_normalized = (arr - arr_min) / (arr_max - arr_min)
print("Mảng sau khi chuẩn hóa:\n", arr_normalized)

# 3. Tính tích vô hướng (dot product) của hai mảng 1D
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
dot_product = np.dot(a, b)
print("Tích vô hướng của hai mảng:", dot_product)

# 4. Tạo ma trận 5x5 và tính định thức và nghịch đảo
matrix = np.random.rand(5, 5)
print("Ma trận:", matrix)
determinant = np.linalg.det(matrix)
print("Định thức của ma trận:", determinant)
if determinant != 0:
    inverse_matrix = np.linalg.inv(matrix)
    print("Ma trận nghịch đảo:", inverse_matrix)
else:
    print("Ma trận không khả nghịch (định thức = 0).")
```

## **Phần 2: Pandas cơ bản**

### **Bài tập 3: Làm quen với DataFrame**
1 Tạo một DataFrame chứa thông tin sau:
| Name    | Age | Score |
|---------|-----|-------|
| Alice   | 23  | 85    |
| Bob     | 25  | 90    |
| Charlie | 22  | 78    |
| David   | 24  | 92    |
| Eva     | 21  | 88    |

2 Tính giá trị trung bình của cột "Score".\
3 Lọc các hàng có "Score" lớn hơn 85.

```python

import pandas as pd

# 1. Tạo DataFrame

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Age': [23, 25, 22, 24, 21],
        'Score': [85, 90, 78, 92, 88]}

df = pd.DataFrame(data)

print(df)

# 2. Tính giá trị trung bình của cột "Score"
mean_score = df['Score'].mean()
print(f"Giá trị trung bình của cột 'Score' là: {mean_score}")

# 3. Lọc các hàng có "Score" lớn hơn 85
filtered_df = df[df['Score'] > 85]
print(filtered_df)
```

## **Phần 3: Làm sạch dữ liệu**

### **Bài tập 5: Xử lý dữ liệu thiếu**
1 Tạo một DataFrame chứa các giá trị sau:

| Name    | Age | City     | Salary |
|---------|-----|----------|--------|
| Alice   | 23  | New York | 60000  |
| Bob     | NaN | Boston   | 52000  |
| Charlie | 25  | NaN      | NaN    |
| David   | 24  | Chicago  | 58000  |
| Eva     | 22  | Boston   | NaN    |

2 Điền giá trị thiếu trong cột Age bằng giá trị trung bình.\
3 Xóa các hàng có nhiều hơn 1 giá trị thiếu.\
4 Điền giá trị thiếu trong cột Salary bằng 50000.

```python
import pandas as pd

# Bước 1: Tạo DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [23, None, 25, 24, 22],  # Giá trị thiếu được biểu diễn bằng None
    "City": ["New York", "Boston", None, "Chicago", "Boston"],
    "Salary": [60000, 52000, None, 58000, None]
}

df = pd.DataFrame(data)

# Bước 2: Điền giá trị thiếu trong cột Age bằng giá trị trung bình
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Bước 3: Xóa các hàng có nhiều hơn 1 giá trị thiếu
df = df.dropna(thresh=df.shape[1] - 1)  # Giữ lại các hàng có ít nhất 3 giá trị không thiếu

# Bước 4: Điền giá trị thiếu trong cột Salary bằng 50.000
df["Salary"] = df["Salary"].fillna(50000)

# Hiển thị DataFrame sau khi xử lý
print(df)
```

## **Phần 4: Trực quan hóa dữ liệu với Matplotlib**

### **Bài tập 6: Biểu đồ cơ bản**

1 Tạo một biểu đồ đường biểu diễn hàm số y = x ** 2 trên khoảng [-10,10] \
2 Vẽ biểu đồ cột thể hiện điểm số (Score) của các sinh viên từ Bài tập 3.\
3 Tạo một biểu đồ tròn (pie chart) thể hiện phần trăm mỗi loại hoa trong tập dữ liệu Iris.

```python

import matplotlib.pyplot as plt
import numpy as np

# 1. Vẽ biểu đồ hàm số y = x^2
x = np.linspace(-10, 10, 100)  
y = x ** 2  # Tính y = x^2

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='y = x^2')
plt.title('Đồ thị hàm số y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

# 2. Biểu đồ cột điểm số
#Code here
student_data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Score": [85, 90, 78, 92, 88]
}
df = pd.DataFrame(student_data)

plt.figure(figsize=(10, 6))
plt.bar(df['Name'], df['Score'])
plt.title('Điểm số của sinh viên')
plt.xlabel('Tên sinh viên')
plt.ylabel('Điểm số')
plt.xticks(rotation=45)
plt.show()

# 3. Vẽ biểu đồ tròn cho dữ liệu Iris
iris_df = pd.read_csv('Iris.csv')
species_counts = iris_df['Species'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(species_counts.values, labels=species_counts.index, autopct='%1.1f%%')
plt.title('Phân bố các loại hoa Iris')
plt.show()
```

### **Bài tập 7: Biểu đồ nâng cao**

1 Vẽ biểu đồ phân tán (scatter plot) giữa sepal_length và sepal_width của tập dữ liệu Iris. Dùng màu sắc để phân biệt các loại hoa (species).\
2 Thêm tiêu đề, nhãn trục và chú thích cho biểu đồ.

```python

# 1. Biểu đồ phân tán với màu sắc theo loại hoa
#Code here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Đọc tập dữ liệu Iris
iris_df = pd.read_csv('Iris.csv') 

# 1. Vẽ biểu đồ phân tán (scatter plot)
plt.figure(figsize=(10, 6))

species = iris_df['Species'].unique()
colors = ['red', 'blue', 'green']  

for specie, color in zip(species, colors):
    subset = iris_df[iris_df['Species'] == specie]
    plt.scatter(
        subset['SepalLengthCm'], 
        subset['SepalWidthCm'], 
        label=specie, 
        color=color,
        alpha=0.7, 
        edgecolor='k'
    )

# 2. Thêm tiêu đề, nhãn trục và chú thích
plt.title('Biểu đồ phân tán giữa Sepal Length và Sepal Width', fontsize=14)
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)
plt.legend(title="Loại hoa (Species)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
